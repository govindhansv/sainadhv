```python
import matplotlib.pyplot as plt

days = [1,2,3,4,5]
sleeping = [10,20,30,40,50]
eating = [20,30,40,50,60]
working = [30,40,50,60,70]  

plt.stackplot(days, sleeping, eating, working, labels=['Sleep','Eat','Work'])
plt.legend(loc='upper left')
plt.title('Stacked Area Plot')
```
